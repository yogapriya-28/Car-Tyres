#**Demand for Tube vs. Tubeless tyres**




```python
# count of products by Resolution
df['Model'].groupby(df['Type']).count().sort_values(ascending=False)
```

```python
# Tyre Type distribution
labels = 'Tubeless', 'Tube'
sizes = [4061,289]
fig, ax = plt.subplots(figsize=(6, 5))
fig.set_facecolor('white')
ax.pie(sizes, labels=labels, colors=["goldenrod", "gold"],autopct='%1.1f%%',startangle=60,textprops={'color':'black','fontsize':12.5})
ax.axis('equal')
plt.show()
```

#**Major players in the market**



```python
df['Model'].groupby(df['Tyre Brand']).count().sort_values(ascending=False)
```

#**Which are the top 5 brands with most tyre models?**





```python
sns.set_style('white')
plt.figure(figsize=(10,7))
sns.countplot(
    y="Tyre Brand",
    hue="Type",
    data=df,
    palette = ["teal", "coral"],
    order=df["Tyre Brand"].value_counts().iloc[:5].index
)

plt.show()
```

#**Top 5 Models with most tyre brands**



```python
sns.set_style('white')
plt.figure(figsize=(10,7))
sns.countplot(
    y="Model",
    hue="Type",
    data=df,
    palette = ["steelblue", "darkorange"],
    order=df["Model"].value_counts().iloc[:5].index
)
plt.title("Top 5 Tyre Brands by Type", fontsize=14)
plt.show()
```

#**Which Maruti model has the highest number of tyre models/serial no. ?**

```python
sns.set_style('white')
plt.figure(figsize=(20,7))

sns.countplot(
    y="Model",
    hue="Tyre Brand",
    data=df,
    palette=sns.color_palette("Set1"),
    order=df["Model"].value_counts().iloc[:1].index
)

plt.title("Top Model by Tyre Brand", fontsize=16)
plt.show()
```

#**Do expensive tyres have higher ratings?**



```python
round(df.groupby('Tyre Brand')['Rating'].mean().sort_values(ascending=False),1)
```

```python
#filter dataframe
df_p=df[(df["Rating"]>= 4.0) & (df["Selling Price"]>= 5000)]
display(df_p.iloc[:5])
```

```python
# Rating vs Tyre Brand
fig, ax = plt.subplots(figsize=(15,6))
ax = sns.barplot(y='Selling Price', x='Tyre Brand', data=df_p, hue='Rating')
plt.show()
```

#**Average Selling Price by Tyre Brand**



```python
df.groupby('Tyre Brand')['Selling Price'].mean().sort_values(ascending=False).apply(np.ceil)
```

```python

data = {
    "Tyre Brand": ["Pirelli", "Michelin", "Maxxis", "Yokohama", "Continental", "BridgeStone",
                   "Falken", "Hankook", "Kumho", "Apollo", "UltraMile", "CEAT", "GoodYear",
                   "MRF", "JKTyre", "Firestone"],
    "Original Price": [7359, 6893, 6232, 6120, 5250, 5190, 5066, 4980, 4650, 4631, 4475,
                       4130, 4011, 3718, 3602, 3484]
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Tyre Brand", y="Original Price", data=df, palette="viridis")
plt.title("Original Price of Tyres by Brand", fontsize=16)
plt.xlabel("Tyre Brand", fontsize=12)
plt.ylabel("Original Price ()", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
